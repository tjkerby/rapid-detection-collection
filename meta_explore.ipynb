```python
rapids = df[df.rapid_class == 1]
rapids.head(10)
```

```python
images = df.image
rapids_len = len(rapids)
a = np.array(["carl", "christy", "julie"])
partition = rapids_len / len(a)

for i in range(rapids_len):
    # print(images[i])
    name_i = int(np.floor(i / partition))
    # print(a[name_i])

    df.loc[(df.image == images[i]), ["user"]] = a[name_i]
df.head(10)
```
